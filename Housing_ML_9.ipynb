## Year Constructed

```python
df.shape
```

```python
pd.set_option('display.max_rows', None)

value_counts_all_YC = df['yearConstructed'].value_counts(dropna=False)
print(value_counts_all_YC)
```

```python
# Replacing missing values in 'yearConstructed' with the median instead of the mean.
# The median is more robust to outliers, as there are extreme values like 1111, 1000, and even 2090 (future).
# The mean could be skewed by these anomalies, leading to a less representative imputation.
# Since construction years tend to cluster in certain periods (e.g., 20th and 21st centuries),
# the median provides a better central tendency measure in this case.
year_reference = 2020

df["yearConstructed"] = df.groupby("geo_plz")["yearConstructed"].transform(lambda x: x.fillna(x.median()))

df["yearConstructed"] = df["yearConstructed"].fillna(df["yearConstructed"].median())

df = df[(df["yearConstructed"] >= 1800) & (df["yearConstructed"] <= 2020)]

df["buildingAge"] = year_reference - df["yearConstructed"]

df = df.drop(columns=["yearConstructed"])
```

## Firing Types

```python
value_counts_all_FT = df['firingTypes'].value_counts(dropna=False)
value_counts_all_FT
```

```python
pd.set_option('display.max_rows', None)

unique_all_FT = df['firingTypes'].unique()
unique_all_FT
```

```python
def categorize_firing_type(value):
    if pd.isna(value):
        return "other_unknown"
    if "electricity" in value:
        return "electricity"
    if "gas" in value:
        return "gas"
    if "oil" in value:
        return "oil"
    if "district_heating" in value:
        return "district_heating"
    if any(x in value for x in ["solar_heating", "geothermal", "wind_energy", "hydro_energy"]):
        return "renewable"
    if any(x in value for x in ["wood", "pellet_heating", "bio_energy", "coal", "wood_chips"]):
        return "biomass/solid_fuel"
    if "combined_heat_and_power" in value:
        return "combined_systems"
    return "other_unknown"

df["firingTypes_grouped"] = df["firingTypes"].apply(categorize_firing_type)
```

```python
pd.set_option('display.max_rows', None)

value_counts_all_FTG = df['firingTypes_grouped'].value_counts(dropna=False)
value_counts_all_FTG
```

## Base Rent

```python
df.shape
```

```python
df['price_m2'] = df['baseRent'] / df['livingSpace']
```

```python
Q1 = df['price_m2'].quantile(0.25)
Q3 = df['price_m2'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['price_m2'] >= lower_bound) & (df['price_m2'] <= upper_bound)]
```
